```python

def valor_por_producto(inventario):
    total_general = 0
    for producto in inventario:
        valor = producto["precio"] * producto["cantidad"]
        print(f"{producto['nombre']}: {valor}€")
        total_general += valor

    print(f"Total inventario: {total_general}€")
```
